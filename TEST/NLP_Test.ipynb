```python
story = """In the rugged terrain of the Western Ghats, Shivaji Maharaj, the founder of the Maratha Empire, was preparing for another day of strategic planning and military campaigns. Known for his innovative tactics and leadership, Shivaji Maharaj was determined to strengthen his kingdom and protect his people from external threats.

Shivaji Maharaj’s stronghold was the Raigad Fort, a formidable fortress perched high on a hilltop. The fort's strategic location allowed him to oversee the surrounding lands and defend against invaders. His trusted advisors and generals would gather in the fort's grand hall to discuss plans, share intelligence, and make critical decisions for the empire's expansion.

The evenings in Shivaji Maharaj's court were vibrant with activity. He would host feasts and gatherings to bolster the morale of his soldiers and celebrate victories. During these events, Shivaji Maharaj would address his people, sharing his vision for a prosperous and united Maratha Empire. His leadership inspired loyalty and courage among his subjects.

As night fell, Shivaji Maharaj would retire to his quarters, reflecting on the day's achievements and preparing for the challenges ahead. His dreams were filled with visions of a strong and resilient empire, where justice and valor would prevail."""
```

```python
import nltk

def tokenize_story(story):
    sentences = nltk.sent_tokenize(story)
    words = nltk.word_tokenize(story)
    return sentences, words

sentences, words = tokenize_story(story)

print("Sentences:", sentences)
print("Words:", words)
```

```python
def lowercase_story(story):
    return story.lower()

lowercase_text = lowercase_story(story)

print(lowercase_text)
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

def remove_stopwords(story):
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(story)
    
    filtered_story = [word for word in word_tokens if word.lower() not in stop_words]
    removed_stopwords = [word for word in word_tokens if word.lower() in stop_words]
    
    return ' '.join(filtered_story), removed_stopwords

filtered_story, removed_stopwords = remove_stopwords(story)

print("Filtered Story:", filtered_story)
print("Removed Stopwords:", removed_stopwords)
```

```python
import string

def remove_punctuation(story):
    return story.translate(str.maketrans('', '', string.punctuation))

cleaned_text = remove_punctuation(story)

print(cleaned_text)
```

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

def stem_story(story):
    ps = PorterStemmer()
    word_tokens = word_tokenize(story)
    
    stemmed_words = [ps.stem(word) for word in word_tokens]
    
    return ' '.join(stemmed_words), stemmed_words

stemmed_story, stemmed_words = stem_story(story)

print("Stemmed Story:", stemmed_story)
print("Stemmed Words:", stemmed_words)
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

def lemmatize_story(story):
    lemmatizer = WordNetLemmatizer()
    word_tokens = word_tokenize(story)
    
    lemmatized_words = [lemmatizer.lemmatize(word) for word in word_tokens]
    
    return ' '.join(lemmatized_words), lemmatized_words

lemmatized_story, lemmatized_words = lemmatize_story(story)

print("Lemmatized Story:", lemmatized_story)
print("Lemmatized Words:", lemmatized_words)
```

```python
import spacy

nlp = spacy.load('en_core_web_sm')

def named_entity_recognition(story):
    doc = nlp(story)
    entities = [(ent.text, ent.label_) for ent in doc.ents]
    return entities

entities = named_entity_recognition(story)

print("Named Entities:", entities)
```

```python
from nltk.tokenize import word_tokenize
nltk.download('averaged_perceptron_tagger')

def pos_tag_story(story):
    word_tokens = word_tokenize(story)
    pos_tags = nltk.pos_tag(word_tokens)
    return pos_tags

pos_tags = pos_tag_story(story)


for word, tag in pos_tags:
    print(f"{word}: {tag}")
```

```python
def term_frequency(word, story):
    word_tokens = word_tokenize(story)
    total_words = len(word_tokens)
    word_count = word_tokens.count(word)
    tf = word_count / total_words
    return tf

tf = term_frequency("Shivaji", story)

print(f"Term Frequency of 'Shivaji': {tf}")
```

```python
from nltk.probability import FreqDist
import matplotlib.pyplot as plt

def plot_frequency_distribution(story):
    word_tokens = word_tokenize(story)
    words = [word for word in word_tokens if word.isalnum()]
    fdist = FreqDist(words)
    top_words = fdist.most_common(10)
    fdist.plot(10, title="Top 10 Most Common Words in the Story")
    
    return top_words

top_words = plot_frequency_distribution(story)

print("Top 10 Most Common Words:", top_words)
```

```python
def count_tokens(story):
    word_tokens = word_tokenize(story)
    return len(word_tokens)

total_tokens = count_tokens(story)

print(f"Total Number of Tokens: {total_tokens}")
```

```python
from nltk.tokenize import sent_tokenize

def count_sentences(story):
    sentences = sent_tokenize(story)
    return len(sentences)

total_sentences = count_sentences(story)

print(f"Total Number of Sentences: {total_sentences}")
```

```python
def stopword_frequency(story):
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(story)
    
    stopword_list = [word for word in word_tokens if word.lower() in stop_words]
    fdist = nltk.FreqDist(stopword_list)
    top_stopwords = fdist.most_common(3)
    
    return top_stopwords, fdist

top_stopwords, fdist = stopword_frequency(story)

print("Top 3 Most Frequent Stopwords:", top_stopwords)
```

```python
def longest_word(story):
    word_tokens = word_tokenize(story)
    longest = max(word_tokens, key=len)
    return longest, len(longest)

longest, length = longest_word(story)

print(f"Longest Word: {longest}")
print(f"Length: {length}")
```

```python
from nltk.tag import pos_tag

def identify_nouns(story):
    word_tokens = word_tokenize(story)
    pos_tags = pos_tag(word_tokens)
    
    nouns = [word for word, pos in pos_tags if pos in ['NN', 'NNS', 'NNP', 'NNPS']]
    
    return nouns


nouns = identify_nouns(story)

print("Nouns in the Story:", nouns)
```

```python
def identify_verbs(story):
    word_tokens = word_tokenize(story)
    pos_tags = pos_tag(word_tokens)

    verbs = [word for word, pos in pos_tags if pos.startswith('VB')]
    
    return verbs

verbs = identify_verbs(story)

print("Verbs in the Story:", verbs)
```

```python
def find_sentences_with_word(story, word):
    sentences = sent_tokenize(story)
    word_sentences = [sentence for sentence in sentences if word.lower() in sentence.lower()]
    return word_sentences

sentences_with_empire = find_sentences_with_word(story, "empire")

print("Sentences with the word 'empire':")
for sentence in sentences_with_empire:
    print(sentence)
    print()
```

```python
nlp = spacy.load('en_core_web_sm')

def categorize_named_entities(story):
    doc = nlp(story)
    people = []
    locations = []
    organizations = []
    
    for ent in doc.ents:
        if ent.label_ == 'PERSON':
            people.append(ent.text)
        elif ent.label_ in ['GPE', 'LOC']:
            locations.append(ent.text)
        elif ent.label_ == 'ORG':
            organizations.append(ent.text)
    
    return people, locations, organizations

people, locations, organizations = categorize_named_entities(story)

print("People:", people)
print("Locations:", locations)
print("Organizations:", organizations)
```

```python
def word_frequency_comparison(story, words):
    word_tokens = word_tokenize(story.lower())
    frequency = {word: word_tokens.count(word) for word in words}
    return frequency

words_to_compare = ["kingdom", "feast"]
frequency = word_frequency_comparison(story, words_to_compare)

print(f"Frequency of 'kingdom': {frequency['kingdom']}")
print(f"Frequency of 'feast': {frequency['feast']}")
```
