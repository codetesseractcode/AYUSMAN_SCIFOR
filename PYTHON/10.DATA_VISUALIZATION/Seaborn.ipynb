## 2. Box Plot

## 2.1 Manual Values

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

oak_heights = [15, 18, 20, 22, 25]
pine_heights = [12, 14, 16, 19, 21]
maple_heights = [17, 19, 21, 23, 26]
birch_heights = [14, 16, 18, 20, 22]

all_heights = oak_heights + pine_heights + maple_heights + birch_heights
species_labels = ['Oak'] * 5 + ['Pine'] * 5 + ['Maple'] * 5 + ['Birch'] * 5

df = pd.DataFrame({'Species': species_labels, 'Height (meters)': all_heights})

plt.figure(figsize=(8, 6))
sns.boxplot(x='Species', y='Height (meters)', data=df)
plt.title("Tree Heights Box Plot")
plt.xlabel("Tree Species")
plt.ylabel("Height (meters)")
plt.grid()
plt.show()
```

# 3. Violin Plot

# 3.1 Manual Dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Category': ['A', 'A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'B'],
    'Value': [1.1, 1.3, 1.5, 1.7, 1.9, 2.1, 2.3, 2.5, 2.7, 2.9]
}

df = pd.DataFrame(data)

plt.figure(figsize=(10,6))
sns.violinplot(x='Category', y='Value', data=df)
plt.title('Violin Plot')
plt.show()
```

## 3.2 CSV Dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("D:\\tips.csv")

sns.violinplot(x="day",y="total_bill", data=df, hue="time")
plt.show()
```

## 3.3 Dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = sns.load_dataset("tips")

sns.violinplot(x="time",y="total_bill", data=df,
               order=["Dinner","Lunch"], inner="stick")
plt.show()
```
